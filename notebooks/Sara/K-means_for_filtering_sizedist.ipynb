```python
df = dsmf['particle_number_size_distribution'].to_pandas().T
```

```python
dfdn = df.dropna()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaled_df = StandardScaler().fit_transform(dfdn)
```

```python
#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 1,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_df)
    sse.append(kmeans.inertia_)

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.ylim([0,1e5])
plt.show()
```

```python
kmeans = KMeans(init="random", n_clusters=8, n_init=10, random_state=1)

#fit k-means algorithm to data
kmeans.fit(scaled_df)

#view cluster assignments for each observation
kmeans.labels_

```

```python
#append cluster assingments to original DataFrame
dfdn['cluster'] = kmeans.labels_

#view updated DataFrame
```

```python
dsmf['cluster'] = dfdn['cluster'].to_xarray()
```

```python
ds_gb = dsmf.groupby('cluster').median()
```

```python

dsmf.groupby('cluster').count()['particle_number_size_distribution'].sel(cluster=1).mean().values
```

```python
for g in ds_gb.cluster:
    counts = dsmf.groupby('cluster').count()['particle_number_size_distribution'].sel(cluster=g).mean().values
    
    ds_gb.sel(cluster=g)['particle_number_size_distribution'].plot(xscale='log', label=f'{g.values}, count: {counts}')
plt.legend()
plt.yscale('log')
```

```python
for g in ds_gb.cluster:
    if g  not in [0,6,7]:
        continue
    counts = dsmf.groupby('cluster').count()['particle_number_size_distribution'].sel(cluster=g).mean().values
    
    ds_gb.sel(cluster=g)['particle_number_size_distribution'].plot(xscale='log', label=f'{g.values}, count: {counts}')

plt.legend()
plt.yscale('log')
```
