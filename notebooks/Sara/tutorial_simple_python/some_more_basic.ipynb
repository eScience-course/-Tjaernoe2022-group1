## Global versus local variables

Next couple of cells are borrowed from: [https://www.w3schools.com/python/python_variables_global.asp](https://www.w3schools.com/python/python_variables_global.asp)

Variables that are created outside of a function (as in all of the examples above) are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.



```python
x = "awesome"

def myfunc():
  print("Python is " + x)

myfunc()
```

```python
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

```python
print(x)
x
```

```python
x = "awesome"

def myfunc():
  global x
  x = "fantastic"
    
    
print("Python is " + x)
myfunc()

print("Python is " + x)
```

## Some tips on organising your code

With time you will get better and better at organising your code. 

#### What does it mean to organise well? 


This is highly subjective, but some overall things to follow: 
- Should be easy to understand what you have done. 
- Data should be stored in a different folder from your code
- Code should be commented well -- your will be shocked at how much time you can waste not understanding your own stuff
- **Clean up your notebooks regularly**

I do it as follows (roughly):
- I make sure one notebook mainly has one task, e.g. one type of plot
- I number my notebooks, sometimes with 1.1., 1.2, 2.1 etc. indicating the order in which they are run. 
- If a notebook reads some data and then writes some data out again, I try to keep track of this by writing all the file names, plus what the notebook does to your data in text, at the top

## Functions and creating your own packages

When you notice you are doing something more than, say, 3 times in a code, it might be time to make it into a function. 

```python
def introduce_me(my_name, time_since_phD, times_at_eScienc):
    out_string = f'Hi! My name is {my_name} and I ' \
        f'am {time_since_phD:.1f} acedemic years old' \
        f" I've been at the eScience course {times_at_eScienc} times"
    return out_string
 
print(introduce_me('Sara', 1.44,4))
```

### Adding default values; 

```python
def introduce_me(my_name='Sara', time_since_phD=1.44, times_at_eScienc=4):
    out_string = f'Hi! My name is {my_name} and I ' \
        f'am {time_since_phD:.1f} acedemic years old' \
        f" I've been at the eScience course {times_at_eScienc} times"
    return out_string
 
print(introduce_me(times_at_eScienc=2))
```
